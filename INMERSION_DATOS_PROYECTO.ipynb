```python
inmuebles.info()
# el precio sigue siendo un objeto, todavia no tenemos un valor numerico. Ahora debemos eliminar el punto
```

```python
inmuebles['Precio'] = inmuebles['Precio'].str.replace('.', '', regex=True) # con la función .replace remplazamos el primer parámetro (un . en este caso) por el segundo (nada en este caso)
# no olvidar el tercer parámetro, ya que así lo exige el lenguaje Python.
```

```python
inmuebles[['Precio', 'Barrio']]
```

```python
# con .astype cambiamos el tipo de dato, de objeto a float por ejemplo.
# Creamos una nueva variable, con el nombre de millon, ya que vamos a restarle 6 ceros al número, para poder tener una dataset no tan emborrosa
inmuebles['Precio_Millon'] = inmuebles.Precio.astype('float')/1000000 # lo dividimos por ese número para reducir 6 ceros
inmuebles.info()
```

```python
# punto 2
# forma 1: primero obtenemos el precio de forma individual de todo nuestro dataset y lo dividimos por el area total del dataset
inmuebles['Valor_m2_Millon'] = inmuebles['Precio_Millon']/inmuebles['Area']
inmuebles.head()
```

```python
#agrupamos por barrio y traemos el promedio de todos los datos numéricos de los barrios
inmuebles.groupby('Barrio').mean()
```

```python
datos_barrio = inmuebles.groupby('Barrio').sum()
datos_barrio
```

```python
# creamos un nuevo atributo (Valor_m2_barrio) dentro de este nuevo data frame (datos_barrio)
datos_barrio['Valor_m2_Barrio'] = datos_barrio['Precio_Millon']/ datos_barrio['Area']
datos_barrio.head()
# en el anterior hizo un promedio de la suma solo de los datos dentro de Precio_Millon y de Area,
# ahora si, suma todos los datos, dentro del barrio, y hace un promedio más correcto.
```

```python
# ahora para agregar estos valores al dataset principal (inmuebles), hacemos un mapeo y almacemos los datos con un diccionario en una nueva variable
m2_barrio = dict(datos_barrio['Valor_m2_Barrio']) # creamos un diccionario en una nueva variable
```

```python
inmuebles['Valor_m2_Barrio'] = inmuebles['Barrio'] # creamos una nueva columna con los mismos valores de Barrio
inmuebles['Valor_m2_Barrio']  = inmuebles['Valor_m2_Barrio'].map(m2_barrio) # reemplazamos los valores por los obtenidos en el diccionario
inmuebles.head(6)
```

```python
inmuebles['Barrio'].value_counts()[:10]
# obtenemos los barrios que más inmuebles tienen a la venta, con su respectivo número de inmuebles disponibles
```

```python
# una nueva variable con, solamente, los nombres ( para ello utilizamos .index) de los barrios con más ventas disponibles
top_barrios = inmuebles['Barrio'].value_counts()[:10].index
```

```python
datos_barrio.reset_index(inplace=True) #con 'inplace = True' puedo realizar la ejecusión de inmediato, sin asignar una nueva variable.
datos_barrio
```

```python
datos_barrio.query('Barrio in @top_barrios') # con @variable me devuelve, en este caso, el top 10 de los barrios
```

```python
#generamos un grafico de barras con seaborn
plt.figure(figsize=(10,8))
ax = sns.barplot(x='Barrio', y='Valor_m2_Barrio', data = datos_barrio.query('Barrio in @top_barrios') )
ax.tick_params(axis='x', rotation=45) #ajustamos los indicadores (giramos los nombres para que sean más legibles)
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x='Barrio', y='Valor_m2_Millon', data = inmuebles.query('Barrio in @top_barrios') )
ax.tick_params(axis='x', rotation=45)
plt.show()
# en esta gráfica podemos ver la mediana de los precios y los outliers, que son los precios que están por debajo y por encima del promedio
```

```python
# achicamos el gráfico con filtros
plt.figure(figsize=(10,8))
ax = sns.boxplot(x='Barrio', y='Valor_m2_Millon', data = inmuebles.query('Barrio in @top_barrios & Valor_m2_Millon < 12') )
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x='Barrio', y='Area', data = inmuebles.query('Barrio in @top_barrios & Area < 300') )
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x='Barrio', y='Precio_Millon', data = inmuebles.query('Barrio in @top_barrios & Precio_Millon < 2000') )
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
# importamos la biblioteca "scikit learn" de Python. ideal para clasificación, regresión, clusterización, reducción de dimensionalidad, selección de modelos, preprocesamiento, entre otros.
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split # importamos paquete que nos permita hacer separación entre modelos
#ambos son paquetes básicos cuando trabajamos con Machine Learning

X = datos_ml[['COD_UPZ_GRUPO']]
y = datos_ml['Precio_Millon']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=99)
# con random_state establezco un estado de aleatoriedad para garantizar la reproductibilidad de mis datos, puedo colocar 99 o cualquier otro número
# con test_size=0.25 le decimos a nuestro modelo que entrene con el 75% porciento de nuestros datos y que vamos a probar con el 25% restante
```

```python
# ahora viene el machine learning, como tal. Con el ML instanciamos un modelo, ajustamos los datos y predecimos
modelo = LinearRegression() # instanciamos una regresión lineal de forma genérica, sin modificar parámetros
```

```python
# tomamos nuestro modelo y hacemos un ajuste de nuestros datos de entrenamientos (train: x e y)
modelo.fit(X_train,y_train)
```

```python
# ahora vemos la prediccion, con nuestros mis datos de x, predice los datos de y
# modelo.predict(X_test). La almacenaremos en una variable
y_predict_test = modelo.predict(X_test)
```

```python
# ahora viene la parte de Validación, dónde utilizamos 2 metricas: error absoluto medio y r2 (r cuadrado), dónde nuestro resultado va de 0 a 1, dónde 1 es correcto y 0 erroneo
from sklearn.metrics import mean_absolute_error, r2_score

baseline_mae = mean_absolute_error(y_test, y_predict_test)
baseline_r2 = r2_score(y_test, y_predict_test)
print(baseline_mae,baseline_r2)
```

```python
# agregamos a la iteración más columnas, antes solamente estaba COD_UPZ
X = datos_ml[['COD_UPZ_GRUPO','Habitaciones','Banos','CONJUNTO_CERRADO','SALARIO_ANUAL_MILLON','TIENE_ESCRITURA']]

y = datos_ml["Precio_Millon"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 99)
modelo_1 = LinearRegression()
modelo_1.fit(X_train, y_train)
y_predict_test = modelo_1.predict(X_test)
y_predict_train = modelo_1.predict(X_train)
mae_test = mean_absolute_error(y_test, y_predict_test)
r2_test = r2_score(y_test, y_predict_test)
mae_train = mean_absolute_error(y_train, y_predict_train)
r2_train = r2_score(y_train, y_predict_train)
print(mae_test,r2_test)
print(mae_train,r2_train)
```
